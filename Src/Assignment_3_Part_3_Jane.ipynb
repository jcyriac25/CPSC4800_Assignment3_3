```python
hdf['Id'] = hdf['Id'].astype(object)
hdf['MSSubClass'] = pd.Categorical(hdf['MSSubClass'])
hdf['MSZoning'] = pd.Categorical(hdf['MSZoning'])
hdf['Street'] = pd.Categorical(hdf['Street'])
hdf['Alley'] = pd.Categorical(hdf['Alley'])
hdf['LotShape'] = pd.Categorical(hdf['LotShape'])
hdf['LandContour'] = pd.Categorical(hdf['LandContour'])
hdf['Utilities'] = pd.Categorical(hdf['Utilities'])
hdf['LandSlope'] = pd.Categorical(hdf['LandSlope'])
hdf['BldgType'] = pd.Categorical(hdf['BldgType'])
hdf['LotConfig'] = pd.Categorical(hdf['LotConfig'])
hdf['HouseStyle'] = pd.Categorical(hdf['HouseStyle'])
hdf['OverallQual'] = pd.Categorical(hdf['OverallQual'])
hdf['OverallCond'] = pd.Categorical(hdf['OverallCond'])
hdf['RoofStyle'] = pd.Categorical(hdf['RoofStyle'])
hdf['RoofMatl'] = pd.Categorical(hdf['RoofMatl'])
hdf['Exterior1st'] = pd.Categorical(hdf['Exterior1st'])
hdf['Exterior2nd'] = pd.Categorical(hdf['Exterior2nd'])
hdf['ExterQual'] = pd.Categorical(hdf['ExterQual'])
hdf['ExterCond'] = pd.Categorical(hdf['ExterCond'])
hdf['Foundation'] = pd.Categorical(hdf['Foundation'])
hdf['MasVnrType'] = pd.Categorical(hdf['MasVnrType'])
hdf['BsmtQual'] = pd.Categorical(hdf['BsmtQual'])
hdf['BsmtCond'] = pd.Categorical(hdf['BsmtCond'])
hdf['BsmtExposure'] = pd.Categorical(hdf['BsmtExposure'])
hdf['BsmtFinType1'] = pd.Categorical(hdf['BsmtFinType1'])
hdf['Heating'] = pd.Categorical(hdf['Heating'])
hdf['HeatingQC'] = pd.Categorical(hdf['HeatingQC'])
hdf['CentralAir'] = hdf['CentralAir'].astype(bool)
hdf['Electrical'] = pd.Categorical(hdf['Electrical'])
hdf['KitchenQual'] = pd.Categorical(hdf['KitchenQual'])
hdf['Functional'] = pd.Categorical(hdf['Functional'])
hdf['FireplaceQu'] = pd.Categorical(hdf['FireplaceQu'])
hdf['GarageType'] = pd.Categorical(hdf['GarageType'])
hdf['GarageFinish'] = pd.Categorical(hdf['GarageFinish'])
hdf['GarageQual'] = pd.Categorical(hdf['GarageQual'])
hdf['GarageCond'] = pd.Categorical(hdf['GarageCond'])
hdf['PavedDrive'] = pd.Categorical(hdf['PavedDrive'])
hdf['PoolQC'] = pd.Categorical(hdf['PoolQC'])
hdf['Fence'] = pd.Categorical(hdf['Fence'])
hdf['MiscFeature'] = pd.Categorical(hdf['MiscFeature'])
hdf['SaleType'] = pd.Categorical(hdf['SaleType'])
```
